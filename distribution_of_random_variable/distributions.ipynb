```python
import numpy as np
import matplotlib.pyplot as plt
```

## Uniform in $(0,1)$

```python
n=100000
U = np.random.rand(n)
_=plt.hist(U,bins=100)
```

```python
n=100000
X=np.random.rand(n)
Y=np.random.rand(n)
Z=X+Y
_=plt.hist(Z,bins=100)
```

```python
n=1000000
Z=np.random.rand(n)+np.random.rand(n)+np.random.rand(n)
_=plt.hist(Z,bins=1000)
```

```python
n=1000000
Z=np.random.rand(n)+np.random.rand(n)+np.random.rand(n)+np.random.rand(n)
_=plt.hist(Z,bins=1000)
```

```python
n=1000000
lambd=1.0
Z=-np.log(np.random.rand(n))/lambd
_=plt.hist(Z,bins=1000,range=(0,50))
```
